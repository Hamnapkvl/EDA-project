Airline Passenger Satisfaction Dataset

This dataset contains survey responses collected from airline passengers regarding their travel experience and overall satisfaction level.
Each row represents a single passenger’s flight experience and includes demographic information, travel details, service quality ratings, operational factors (such as delays), and the final satisfaction outcome.

Business Context

Airlines operate in a highly competitive environment where customer experience directly influences:
Customer retention
Brand reputation
Revenue growth
Loyalty program performance
Understanding what drives passenger satisfaction is critical for strategic decision-making.

Core Business Problem

The airline wants to answer:
What factors most strongly influence passenger satisfaction?
How do delays affect customer perception?
Does travel class justify price differences in satisfaction levels?
Which customer segments are most satisfied or dissatisfied?
Which service areas require improvement?
